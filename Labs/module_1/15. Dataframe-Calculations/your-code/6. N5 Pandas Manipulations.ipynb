# Dummies

If you are planning to do some analysis of this series, there is a chance you would think that title makes impact.  
Well, if we think about series that have 'Marvel' or 'DC' in the title, we already know the platform they use to stream it and the genre. What about the once, containing 'Super' or 'Dragon'?

In order to use titles in our Machine Learning Analysis we need to define a numeric representation for each value we see in the Series. The most popular approach is to use One Hot Encoding approach. 

One Hot Encoding generates a [dummy variable](https://en.wikipedia.org/wiki/Dummy_variable_(statistics)) for each category that you meet in your dataset with values being `0` or `1`, depending whether the value was met.

In Pandas you have `get_dummies()` method that gets dummies :D

There are different implementations of this method. One is done via `str.get_dummies`.

````python
df.todolist.str.get_dummies().head()
````

````python
df.todolist.str.get_dummies(sep=' ').head()
````

As you can see, we obtained separate column for each value we have in the `todolist` column. After specifying a separator `sep` we obtained a separate column for each unique element in column.

Second implementation of this method is done via `pd.get_dummies`. It also can perform differently depending on the purpose of study.

````python
# get dummies for specific columns
pd.get_dummies(df.todolist, drop_first=True).head()
````

Here we created a dummy column for `todolist`. Note the `drop_first` attribute, it drops one of the categories existing in the column. The reason, why we need to use it lays behind some numerical problems we will encounter later. 

For now, you can consider the following logic: 
* 1 stands for `to watch`
* 0 stands for `skip`  
which actually makes a lot of sense. Why do we need to keep both?

````python
# get dummies for specific columns and put them back into df
pd.get_dummies(data=df, columns=['todolist'], prefix='buaa')
````

Second approach can be handy when you need to add your dummy columns back into your dataframe. You can define `prefix` attribute to add prefix to dummy column names. 

# Binning

Sometimes when you work with numeric columns, it makes sense to bin them. When?

Consider the column IMDb. DO you think there is a difference between 9.9 and 10.0? Is this difference is equal to difference between 7.9 and 8.0? Having in mind that those scores are averages, series that was scored as 10 is definitely the best, while others are not. And yeah, human beings usually classify those kind of scores as `Terrible`, `Bad`, `Fine`, `Goodish`, `Amazing`. Those categories may vary per person, but the core of it is still there. How can we bin our column?

````python
#define labels
labels = ['Terrible', 'Bad', 'Fine', 'Goodish', 'Amazing']
````

There are 3 main ways we bin our data:
* Equal width bins: the range for each bin is the same size.
* Equal frequency bins: approximately the same number of records in each bin.
* Custom-sized bins: the user explicitly defines where they want the cutoff for each bin to be.

And yeah, Pandas can implement each of them.

## Equal width bins

````python
bins=pd.cut(df.IMDb, 5, labels=labels)
display(bins.head())

#get numeric codes for categories
display(bins.cat.codes.head())
````

````python
bins, representations=pd.cut(df.IMDb, 5, labels=labels, retbins=True)
display(bins.value_counts())

display(representations)
````

Setting `retbins=True` outputs the numerical representation of each bin. Here, we obtained the following bins:
* (0.99 - 3] - Terrible

```python
def select_correct_df(df, type_of_fuel):
    d1=df.loc[df['Fuel Type']==type_of_fuel]
    d2=d1.groupby('Year')[['Combined MPG', 'City MPG']].mean()
    d3=d2.melt(var_name='MPG type', ignore_index=False)
    return d3
```

```python
fig, axes=plt.subplots(1,2, figsize=(12,6), sharey=True)
for ix, value in enumerate(['Regular','Premium']):
    sns.lineplot(data=select_correct_df(df, value), 
                 y='value',
                 x='Year', 
                 hue='MPG type',
                ax=axes[ix])
    axes[ix].set_title(value+' type of fuel')
axes[1].get_legend().remove()
plt.show()
```

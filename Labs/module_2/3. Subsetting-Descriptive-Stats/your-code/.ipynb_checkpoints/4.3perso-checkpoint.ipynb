```python
#how many cars per year
df.groupby('Year').Model.nunique()
```

```python
df.groupby('Year').Make.agg('count')
```

```python
#actually we need to consider some columns to make it unique
x=df.groupby('Year').Make.agg('count')

plt.plot(x)   #plot is line plot
```

```python
#then you can play with the style (formating)
plt.plot(x,'ro--')
```

```python
#what is inside the plot that we can change
*Data
*Plot type
*Figure size
*Limits for x and y axis
*X-label and Y-label
*title
*Legend
*Border
*Background
*Grid
*Canvas - Figure
```

## Canvas

```python
#you need to put the plt in the same cell as the fig
fig=plt.figure()
plt.plot(x,'bo--')
plt.plot(df['Engine Displacement'])
```

```python
fig=plt.figure()
plt.plot(x,'bo--')
plt.show() #if you put this line it stops the first plot and you start the second one in another one
plt.plot(df['Engine Displacement'])
```

```python
fig=plt.figure()
plt.plot(x,'bo--', label='cars')
plt.title('Very nice plot')
plt.xlabel('Year')
plt.ylabel('Number of Models')
plt.grid()
#plt.axis('off')  # useful for sankey (no need for axis)
plt.legend()
plt.xticks(rotation=120)
plt.show()
```

## Average amount of cylinders per year

```python
x=df.groupby('Year').Cylinders.agg('mean')
plt.plot(x,'bo--')
```

```python
x.values
```

```python
x.index
```

```python
plt.bar(x.index,x.values)
```

```python
#with pandas (automatic without specifying x and y axis)
x.plot.bar()
```

```python
x.plot.bar(label='smth', figsize=(6,3))
#after using x.plot you can use matplotlib after that
plt.legend()
plt.ylabel('boo')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.set('poster', palette='deep')
```

```python
fig=plt.figure()
x.plot.bar(label='smth', figsize=(6,3))
plt.legend()
plt.ylabel('boo')
plt.show()
```

```python
sns.barplot(data=x.reset_index(), x='Year', y='Cylinders')
```

```python
plt.bar(x.index, x.values);
x.plot.bar;
sns.barplot(data=x.reset_index(), x='Year',y='Cylinders');
```

```python
plt.scatter(x=df['Combined MPG'], y=df['Fuel Cost/Year'])
```

```python
#The better version is with seaborn
fig=plt.figure()
sns.scatterplot(data=df, x='Combined MPG', y='Fuel Cost/Year')
plt.title('Smth')
plt.show()
```

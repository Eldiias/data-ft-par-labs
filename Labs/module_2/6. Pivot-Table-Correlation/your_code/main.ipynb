#### What's the mean `Salary` by `Department`?
**Tip**: There are 2 ways to do it: with or without pivot tables. Do it both ways. 

```python
# Method 1
mean_salary1=employees.groupby('Department')['Salary'].agg('mean')
plt.plot(mean_salary1)
```

```python
# Method 2
mean_salary2=pd.pivot_table(data=employees, index='Department',values='Salary',aggfunc='mean')
mean_salary2
```

```python
plt.plot(mean_salary2)
```

## Challenge 1

#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
fig = plt.figure()
plt.legend()
#plot(x,y)
plt.plot(x,y, 'go')
plt.show()

#plot(x,z)
plt.plot(x,z, 'o--')
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(x,z)

plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
matplotlib.rc('font', size=20)
fig, axes = plt.subplots(1,2, figsize=(12,5))

axes[0].set_title('plot 1')
axes[0].plot(x,y)
axes[0].legend("a")

axes[1].set_title('plot 2')
axes[1].plot(x,z)
axes[1].legend("b")
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
fig = plt.figure(figsize=(15,15))
fig, axes = plt.subplots(1,2, figsize=(15,5))
axes[0].plot(x**2, np.exp(x))
axes[0].set_title('Normal')


axes[1].plot(x**2, np.exp(x))
axes[1].set_xscale("log")
axes[1].set_yscale("log") 
axes[1].set_title('LOG')
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig = plt.figure(figsize=(15,15))
fig, axes = plt.subplots(1,2, figsize=(15,5))
axes[0].plot(x**2, np.exp(x))
axes[0].legend('a')

axes[1].plot(x**2, np.exp(x))
axes[1].set_xscale("log")
axes[1].set_yscale("log") 
axes[1].legend("b")
```

#### From the Fitbit data, we want to visually understand:

```python
months = fitbit.groupby('Months_encoded').Steps.median()
months
```

```python
# 0 = weekend
# 1 = week
weekend = fitbit.groupby('Work_or_Weekend').Steps.median().reset_index()
weekend
```

```python
matplotlib.rc('font', size=10)
fig, axes = plt.subplots(1,2, figsize =(12,5))

# MEDIAN STEPS BY MONTH_ENCODED
# your code here

axes[0].plot(months)
axes[0].set_title('Median Steps per month')



# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
axes[1].hist(weekend)
axes[1].set_title('Med.Steps weekend/weekday')

plt.show()
```

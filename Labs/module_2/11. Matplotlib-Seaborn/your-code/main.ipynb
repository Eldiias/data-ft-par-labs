#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1
fig = plt.figure(figsize=(8, 5))
plt.plot(x, y, color='green', linestyle='dashed', linewidth=2)
plt.plot(x,z, color='red', linestyle='-', linewidth=2)
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, color='green', linewidth=2)
axes[1].plot(x,z, color = 'red', linewidth = 3)
plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (8,5))
axes[0].plot(x, y, color='green', linewidth=2)
axes[1].plot(x,z, color = 'red', linewidth = 3)
axes[0].set_title('x-plot')
axes[1].set_title('y-plot')
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[1].set_xlabel('x-axis')
plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (8,5))
axes[0].plot(x, x**2, color='green', linewidth=2)
axes[1].plot(x, np.exp(x), color = 'red', linewidth = 3)
axes[1].set_yscale('log')
axes[0].set_title('x-plot')
axes[1].set_title('y-plot')
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[1].set_xlabel('x-axis')
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (8,5))
axes[0].plot(x, x**2, color='green', linewidth=2, label = 'normal plot')
axes[1].plot(x, np.exp(x), color = 'red', linewidth = 3, label = 'log plot')
axes[1].set_yscale('log')
axes[0].set_title('x-plot')
axes[1].set_title('y-plot')
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[1].set_xlabel('x-axis')
axes[0].legend()
axes[1].legend()
plt.show()
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fig = plt.figure(figsize=(8,4))
monthly = fitbit2.groupby('Months_encoded')['Steps'].median()
monthly.plot.bar( color = 'green') 
plt.ylabel('No of Steps')
plt.title("Monthly Steps' Median")
plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
week = fitbit2.groupby('Work_or_Weekend')['Steps'].median()
week.plot.bar(color = 'orange')
plt.ylabel('No of Steps')
plt.title("Work_or_Weekdays Steps' Median")
plt.show()
```

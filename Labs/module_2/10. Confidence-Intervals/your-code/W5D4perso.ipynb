## Confidence intervals

## when population std is known

```python
var_pop=np.var(df1.Rating)
```

```python
var_pop
```

```python
df1.Rating.var()
```

```python
from scipy.stats import norm
```

```python
We will study 5% error case. Hence alpha/2 = 0.025
```

```python
round(norm.ppf(0.025),2)
```

```python
#first approach with the formula
z_negative=norm.ppf(0.025)
```

```python
first_sample=samples[0]
```

```python
lower=means[0]+z_negative*var_pop/np.sqrt(150)
upper=means[0]-z_negative*var_pop/np.sqrt(150)
```

```python
print('Confidence interval for the first sample is :', round(lower,3), round(upper,3))
```

```python
df1.Rating.mean()
```

```python
#second approach
#loc is sample mean, scale is pop_variance over sqrt(n)
```

```python
norm.interval(0.95, loc=means[0], scale=var_pop/(150**0.5))
```

```python
norm.interval(0.95, loc=means[0], scale=df1.Rating.var(ddof=0)/(150**0.5))
```

```python
def ci_norm(x, sigma_squared=var_pop):
    standard_error=(sigma_squared/x.shape[0])**0.5
    return norm.interval(0.95,scale=standard_error)
```

```python
ds=pd.DataFrame(map(ci_norm,samples), columns=['lower','upper'])
```

```python
ds
```

```python
ds['means']=means
```

```python
ds['population_mean']=df1.Rating.mean()
```

```python
ds
```

```python
ds['inside']=ds.population_mean.between(ds.lower+ds.means,ds.upper+ds.means)
```

```python
ds.eval('lower<(population_mean-means)<upper')
```

```python
ds['inside']=ds.eval('lower<(population_mean-means)<upper')
```

```python
ds.inside.value_counts(normalize=True)
```

```python
For every sample average I check if the mean is inside the IC, here we found 98,8% of the averages inside the confidence interval
```

```python
dds=ds.sample(250)
```

```python
plt.figure(figsize=(20,8))
for inside, x in dds.groupby('inside'):
    plt.errorbar(x.index, x.means, yerr=x.upper.values, fmt='o', uplims=False, lolims=False)
plt.plot(range(len(samples)), dds.population_mean)
plt.show()
```

```python
plt.figure(figsize=(20,8))
for inside, x in ds.groupby('inside'):
    plt.errorbar(x.index, x.means, yerr=x.upper.values, fmt='o', uplims=False, lolims=True)
plt.plot(range(25000), ds.population_mean)
plt.show()
```

## Unknown population variance

```python
from scipy.Stats import t
```

```python
def ci_t(x):
    s_squared=x.Rating.var()
    standard_error=(s_squared/x.shape[0])**0.5
    return norm.interval(0.95,df=x.shape[0]-1, scale=standard_error)
```

```python
ds=pd.DataFrame(map(ci_t,samples), columns=['lower','upper'])
ds['means']=means
ds['population_mean']=df1.Rating.mean()
ds['inside']=ds.eval('lower+means<population_mean<upper+means')
ds.inside.value_counts(normalize=True)
```

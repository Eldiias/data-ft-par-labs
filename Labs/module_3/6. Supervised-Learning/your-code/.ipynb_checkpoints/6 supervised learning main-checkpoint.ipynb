Although there are so many unique values in the `SERVER` column, there are actually only 3 main server types: `Microsoft`, `Apache`, and `nginx`. Just check if each `SERVER` value contains any of those server types and re-label them. For `SERVER` values that don't contain any of those substrings, label with `Other`.

At the end, your `SERVER` column should only contain 4 unique values: `Microsoft`, `Apache`, `nginx`, and `Other`.

```python
df.Server.fillna('Other', inplace = True)
```

```python
df['New_searver'] = df.SERVER.str.contains('apache', case = False)
```

```python
df.loc[df.SERVER.str.contains('apache', case = False), 'New_server'] = 'Apache'
df.loc[df.SERVER.str.contains('microsoft', case = False), 'New_server'] = 'Microsoft'
df.loc[df.SERVER.str.contains('nginx', case = False), 'New_server'] = 'Nginx'
```

```python
df.loc[df.New_server.str.contains==False, 'New_server'] = 'Other'
```

```python
df.drop('SERVER', axis = 1 inplace = True)
```

```python
def server_clean(server):
    brands = ['Microsoft', 'Apache','nginx']
    result = 'Other'
    for b in brands:
        if b in str(server):
            result = b
    return result
```

```python
df.SERVER=df.SERVER.apply(lambda x: server_clean(x))
```

```python
# Count `SERVER` value counts here
df.SERVER.value_counts()
```

Problem-solving in machine learning is iterative. You can improve your model prediction with various techniques (there is a sweetspot for the time you spend and the improvement you receive though). However, now we do want you to learn one of the advanced techniques which is called *feature scaling*. The idea of feature scaling is to standardize/normalize the range of independent variables or features of the data. This can make the outliers more apparent so that you can remove them. This step needs to happen during Challenge 6 after you split the training and test data because you don't want to split the data again which makes it impossible to compare your results with and without feature scaling. For general concepts about feature scaling, click [here](https://en.wikipedia.org/wiki/Feature_scaling). To read deeper, click [here](https://medium.com/greyatom/why-how-and-when-to-scale-your-features-4b30ab09db5e).

In the next cell, attempt to improve your model prediction accuracy by means of feature scaling. A library you can utilize is `sklearn.preprocessing.RobustScaler` ([documentation](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html)). You'll use the `RobustScaler` to fit and transform your `X_train`, then transform `X_test`. You will use logistic regression to fit and predict your transformed data and obtain the accuracy score in the same way. Compare the accuracy score with your normalized data with the previous accuracy data. Is there an improvement?

```python
from sklearn.preprocessing import RobustScaler
```

```python
scaler = RobustScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model4 = LogisticRegression()
model4.fit(X_train_scaled, y_train)
y_pred4 = model4.predict(X_test_scaled)
confusion_matrix(y_test, y_pred4)
```

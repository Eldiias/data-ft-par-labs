```python
from sklearn import datasets
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y=datasets.make_moons(noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
```

# Get the actual data

```python
cancer = datasets.load_breast_cancer()
```

```python
print(cancer.DESCR)
```

```python
import pandas as pd
```

```python
X=pd.DataFrame(cancer.data, columns=cancer.feature_names)
```

```python
X.head()
```

```python
y=cancer.target
```

```python
y
```

```python
X.isna().sum().sum()
```

```python
X.describe().T
```

# Preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
scaler=StandardScaler()
scaler.fit(X_train)
```

```python
X_train_scaled=scaler.transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

# Modeling (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
```

```python
y_pred=knn.predict(X_test_scaled)
```

```python
y_pred
```

```python
X.columns
```

# Evaluation

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
pca.fit(X_train_scaled)
```

```python
pca.explained_variance_ratio_.round(2)
```

```python
X_train_scaled_pca=pca.transform(X_train_scaled)
```

```python
pd.DataFrame(X_train_scaled_pca).corr().round(); #just to show that correlation between different principal components is 0
```

```python
pca=PCA(6)
pca.fit(X_train_scaled)
pca.explained_variance_ratio_.sum()
```

```python
pca=PCA(0.85)
pca.fit(X_train_scaled)
pca.explained_variance_ratio_.sum()
```

```python
X_train_scaled_pca=pca.transform(X_train_scaled)
```

```python
X_train_scaled_pca.shape
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train_scaled_pca, y_train)
```

```python
X_test_scaled_pca=pca.transform(X_test_scaled)
```

```python
y_pred_pca=knn.predict(X_test_scaled_pca)
confusion_matrix(y_test, y_pred_pca)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
plt.scatter(X_train_scaled_pca[:,0], X_train_scaled_pca[:,1], c=y_train)
```

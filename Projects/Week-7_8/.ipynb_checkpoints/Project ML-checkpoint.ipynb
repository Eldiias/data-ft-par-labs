#### Data Set Description

1. Age: individual's age
2. Sex: indvidual's gender
        1 = male
        2 = female
3. Chest Pain Type (cp): type of chest pain individual has
        1 = typical angina
        2 = atypical angina
        3 = non-anginal pain
        4 = asymptomatic
4. Resting Blood Pressure (trestbps): resting blood pressure in mmHg
5. Serum Cholesterol (chol): serum cholesterol in mg/dl
6. Fasting Blood Sugar (fbs): compares individual fasting blood sugar with 120 mg/dl
        1 = True (fasting blood Sugar > 120 mg/dl)
        2 = False (fasting blood sugar < 120 mg/dl)
7. Resting ECG (restecg): Electrocardiogram results
        0 = Normal
        1 = Having ST-T wave abnormality
        2 = left ventricular hypertrophy
8. Maximum Heart Rate Achieeved (thalach): max heart rate by individual
9. Exercise induced angina (exang)
        1 = yes
        2 = no
10. ST Depression induced by exercise relative to rest(oldpeak)
11.  the slope of the peak exercise ST segment (slope):
        0 = upsloping
        1 = flat
        2 = downsloping
12. Number of major vessels (0-3) colored by flourosopy (ca): values from 0 - 4
13. Thalassemia (thal): Blood disorder when blood doesn't make heamoglobin
        3 = normal
        6 = fixed defect
        7 = reversible defect
14. Heart disease diagnosis (target):
        0 = less chance of heart attack
        1 = more chance of heart attack

## Evaluating Supervised Learning Algorithms

### Importing Requisite Libraries

```python
# Import function to split the DataFrame

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Import different models

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Import the scoring
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
#Import cross validation libraries and hyperparameter

from sklearn.model_selection import StratifiedKFold, KFold, RepeatedKFold, RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

### Spliting Data into Training and Testing Sets

```python
X_train, X_test, y_train, y_test = train_test_split(df_dummies.drop('target', axis=1),
                                                    df_dummies.target,
                                                    test_size=0.2,
                                                    stratify=df.target, 
                                                    random_state = 42)
```

#### Scaling Data using MinMaxScaler

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Cross Validation & Hyperparameter Tuning

```python
cv = RepeatedStratifiedKFold(n_splits=20, n_repeats=10, random_state=42)
```

### Support Vector Machine

```python
# Without Hyperparameter tuning and cross validation

model1 = SVC()
model1.fit(X_train, y_train)
y_pred1 = model1.predict(X_test)
print(classification_report(y_test, y_pred1))
```

```python
model_svc= SVC()
params_svc = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel':['rbf', 'linear']}

# Using Gridsearch for hyperparameters
# grid_svc = GridSearchCV(estimator=model_svc, cv = cv, param_grid=params_svc, scoring='accuracy', n_jobs=1)
grid_svc = GridSearchCV(estimator=model_svc, param_grid=params_svc, refit=True, verbose=3)
grid_svc.fit(X_train, y_train)
grid_svc_pred = grid_svc.predict(X_test)
dfsvc = pd.DataFrame(grid_svc.cv_results_)
dfscv = dfsvc[['mean_test_score', 'params']]
```

```python
# printing out best accuracy scores and parameters
print(grid_svc.best_params_)
print(grid_svc.best_score_)

dfscv.head(3)
```

```python
# Without Hyperparameter tuning and cross validation

model2 = LogisticRegression()
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)
print(classification_report(y_test, y_pred2))
```

```python
model_lr = LogisticRegression()
params_lr = {'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
            'multi_class' : ['auto', 'ovr']}

# Using Gridsearch for hyperparameters
grid_lr = GridSearchCV(estimator=model_lr, cv = cv,  param_grid=params_lr, scoring='accuracy', n_jobs=1)
grid_lr.fit(X_train, y_train)
dflr = pd.DataFrame(grid_lr.cv_results_)
dflr = dflr[['mean_test_score', 'params']]

# printing out best accuracy scores and parameters
print(grid_lr.best_params_)
print(grid_lr.best_score_)

dflr.head()
```

### Comparing Models

```python
cv = RepeatedStratifiedKFold(n_splits=25, n_repeats=5)

model_params = {
    'SVC': {'model': SVC(gamma=0.0001),
           'params':{'kernel':['rbf', 'linear'], 'C':[1,10,20]}},
    
    'Logistic Regression':{'model': LogisticRegression(), 
                          'params':{'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
                                    'multi_class' : ['auto', 'ovr']}},
    
    'KNearest Neighbors':{'model': KNeighborsClassifier(n_neighbors=15), 
                          'params':{'weights' : ['uniform', 'distance'],
                                    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute']}},
    
    'Gaussian NB':{'model': GaussianNB(), 
                          'params':{'var_smoothing' :np.logspace(0,-9, num=100)}},
    
    'Random Forest':{'model': RandomForestClassifier(n_estimators=50), 
                          'params':{'max_features' :["auto", "sqrt", "log2"], 
                                    'class_weight' : ["balanced", "balanced_subsample"]}},
                     
    'Decision Tree':{'model': DecisionTreeClassifier(), 
                          'params':{'criterion' : ["gini", "entropy"], 'splitter' : ["best", "random"]}}
}
```

```python
scores = []

for model_name, model_param in model_params.items():
    grid = GridSearchCV(model_param['model'], model_param['params'], cv=cv)
    grid.fit(X_train, y_train)
    scores.append({'Model Name':model_name,
                  'Best Accuracy Score':grid.best_score_,
                  'Best Parameters':grid.best_params_})
```

```python
df_models = pd.DataFrame(scores)
df_models
```

```python
# Plot comparing models

sns.set(style='white')
fig = plt.figure(figsize=(12,8))
model_plot = sns.barplot(data=df_models, x=df_models['Model Name'], y=df_models['Best Accuracy Score'], palette='RdBu')

for p in model_plot.patches:
    model_plot.annotate(format(p.get_height(), '.4f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points', fontsize=14)

plt.title('Classification Model Comparison', fontsize=15)
plt.xlabel(None)
plt.ylabel('Best Accuracy Score', fontsize=14)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14, rotation=35)

plt.savefig('8. Model Comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

## PCA & Kmeans Clustering

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# Preliminary checks to see if there are any cluster

sns.set(style='ticks')
sns.pairplot(data=df[['age','trestbps','chol','oldpeak','thalach']])

plt.show()
```

```python
scale = MinMaxScaler()
```

```python
pca = PCA()
X = df.iloc[:, :-1].values
scaler.fit_transform(X)
pca_df = pca.fit_transform(X)
pca_df.shape
```

```python
# Applying kmeans to create clusters

kms = KMeans(n_clusters=4, max_iter=100, random_state=42)
kms.fit_transform(pca_df)

#labels for clusters
label = kms.fit_predict(pca_df)
```

```python
#filter rows of original data
fig = plt.figure(figsize=(10,7))
filtered_label0 = pca_df[label == 0]
filtered_label1 = pca_df[label == 1]
filtered_label2 = pca_df[label == 2]
filtered_label3 = pca_df[label == 3]
 
#plotting the results
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1])
plt.scatter(filtered_label1[:,0] , filtered_label1[:,1] , color = 'red')
plt.scatter(filtered_label2[:,0] , filtered_label2[:,1] , color = 'black')
plt.scatter(filtered_label3[:,0] , filtered_label3[:,1] , color = 'green')

plt.title('Plot of 4 Clusters (KMeans Clusters)', fontsize = 14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
#Getting unique labels
fig = plt.figure(figsize=(10,7))
u_labels = np.unique(label)
 
#plotting the results:
 
for i in u_labels:
    plt.scatter(pca_df[label == i , 0] , pca_df[label == i , 1] , label = i)
plt.legend(fontsize=14, title='Clusters', title_fontsize='large')
plt.title('Plot of 4 Clusters (KMeans Clusters)', fontsize = 14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
#Getting the Centroids
fig = plt.figure(figsize=(10,7))
centroids = kms.cluster_centers_
u_labels = np.unique(label)
 
#plotting the results:
 
for i in u_labels:
    plt.scatter(pca_df[label == i , 0] , pca_df[label == i , 1] , label = i)
plt.scatter(centroids[:,0] , centroids[:,1] , s = 200, marker='*', color = 'k')
plt.legend()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.title('Plot of 4 Clusters (KMeans Clusters)', fontsize = 14)
plt.savefig('9. KMeans_4_Clusters.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df_km = df.iloc[:, :-1]
df_km.head(2)
```

```python
k_range = range(1,15)
sse = []

for k in k_range:
    kms = KMeans(n_clusters=k)
    kms.fit(df_km)
    sse.append(kms.inertia_)
```

```python
fig = plt.figure(figsize=(10,7))

plt.plot(k_range, sse)
plt.xlabel('k', fontsize = 14)
plt.xticks(fontsize=14)
plt.ylabel('Distortion Score', fontsize=14)
plt.yticks(fontsize=14)
plt.title('Distortion Score Elbow', fontsize = 14)
plt.savefig('11. Elbow1.png', dpi=300, bbox_inches='tight')

plt.show()
```

## Web Scraping Yahoo Finance
The idea here is to scrap financial data from Yahoo Finance website (I chose page the most active stocks) and try to answer a question

```python
# first things first. I found out that the website uses API POST. So no need for beautiful soup.
import requests as r
import time
import pandas as pd
```

```python
link='https://query1.finance.yahoo.com/v1/finance/screener?crumb=00%2Finl.UKdq&lang=en-US&region=US&formatted=true&corsDomain=finance.yahoo.com'
```

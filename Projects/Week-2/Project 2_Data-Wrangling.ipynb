### Missing Values

#### Dropping rows with ALL NaN values
- From the tail, it can bee noticed alot of the rows have no values
    - Thus i dentify all the rows with no values and drop them first
- There are 24 columns, thus identifying rows with at least 20 values allows us to delete these

```python
df = df[df.isnull().sum(axis = 1) < 20]
```

- Next, dropping dates older than 1900 as these dates are not populated

```python
df = df.drop(df.iloc[5559:].index)
```

#### Dropping Columns with all NaN

```python
null_cols = df.isnull().sum()
```

```python
# Analying the percentage of missing values

missing_percentage = round(null_cols[null_cols > 0] / df.shape[0]*100, 2)
missing_percentage
```

- Analyzing the percentage of missing values 
    - From the above, all the columns have missing values though some are quite insignificant at 0.02%
    - Two columns (Unnamed: 22 and Unnamed: 23) both have almost al of their values missing and thus can be dropped

```python
pd.set_option('display.max_rows', None)
```

```python
df = df.drop(['Unnamed: 22', 'Unnamed: 23', 'Date'], axis = 1)
```

- The case number, case number.1 and case number.2 from random samples show to be the same dates, thus data can be used from either of these columns to complete the date column

```python
df['Date'] = df['Case Number.1'].str.extract(r'([0-9]{4}.[0-9]{2}.[0-9]{2})')
```

```python
df['Date'] = df['Date'].str.replace('[.]', '-')
```

- Rearranging the columns to bring forth the New Date column
